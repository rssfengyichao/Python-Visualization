## matplotlib 入门

### 图片与子图

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax1.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
fig, axes = plt.subplots(2,3)
axes[0,1].plot(np.random.randn(50).cumsum(), 'k--')
```

### 颜色、标记和线类型

### 刻度、标签和图例

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(), label='one')
ax.set_xticks([0,250,500,750,1000])
ax.set_xticklabels(['one','two','three','four','five'], rotation=30,fontsize='small')
ax.set_title('Title')
ax.set_xlabel('Stage')
ax.set_ylabel('Y')

ax.plot(np.random.randn(1000).cumsum(), label='two')

ax.legend(loc='best')
```

## 基本图形
### 条形图/柱状图

```python
fig, axes = plt.subplots(2,1)
data = pd.Series(np.random.rand(16), index = range(16))
data.plot.bar(ax = axes[0], color='k', alpha=0.7)
axes[0].set_title('Bar')
data.plot.barh(ax = axes[1], color='k', alpha=0.7)
```
