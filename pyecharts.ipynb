```python
from pyecharts import options as opts

bar = Bar()
bar.add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
bar.add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
bar.set_global_opts(title_opts=opts.TitleOpts(title='主标题',subtitle='副标题'))
bar.render_notebook()
```

```python
from pyecharts.render import make_snapshot

# 使用 snapshot-selenium 渲染图片
from snapshot_selenium import snapshot

bar = (
    Bar()
    .add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
    .add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
)
make_snapshot(snapshot, bar.render(), "bar.png")
```

```python
from pyecharts.globals import ThemeType

bar = Bar(init_opts=opts.InitOpts(theme=ThemeType.LIGHT))
bar.add_xaxis(["衬衫", "羊毛衫", "雪纺衫", "裤子", "高跟鞋", "袜子"])
bar.add_yaxis("商家A", [5, 20, 36, 10, 75, 90])
bar.add_yaxis("商家B", [15, 6, 45, 20, 35, 66])
bar.set_global_opts(title_opts=opts.TitleOpts(title="主标题", subtitle="副标题"))
bar.render_notebook()
```

## 基本图表
### 饼图

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Page, Pie

c = Pie()
c.add("", [list(z) for z in zip(Faker.choose(), Faker.values())],radius=["40%", "70%"])
# c.set_colors(["blue", "green", "yellow", "red", "pink", "orange", "purple"])
c.set_global_opts(title_opts=opts.TitleOpts(title="Pie-基本示例"),
                 legend_opts=opts.LegendOpts(orient="vertical", pos_top="15%", pos_left="2%"))
c.set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
c.render_notebook()
```

### 柱状图/条形图

```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar
from pyecharts.globals import ThemeType

c = Bar()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values())
c.add_yaxis("商家B", Faker.values())
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例", subtitle="我是副标题"))
c.render_notebook()
```

- 通过dict进行配置

```python
c = Bar({"theme": ThemeType.MACARONS})
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values())
c.add_yaxis("商家B", Faker.values())
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例", subtitle="我是副标题"))
c.render_notebook()
```

- 默认取消显示某 Series

```python
c = Bar()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values())
c.add_yaxis("商家B", Faker.values(), is_selected=False)
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例", subtitle="我是副标题"))
c.render_notebook()
```

- 显示 ToolBox
- 不显示图例

```python
c = Bar()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values())
c.add_yaxis("商家B", Faker.values())
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例"),
                 toolbox_opts=opts.ToolboxOpts(),
                legend_opts=opts.LegendOpts(is_show=False))
c.render_notebook()
```

- 单系列柱间距离

```python
c = Bar()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values(), category_gap="80%")
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例"))
c.render_notebook()
```

- 不同系列柱间距离

```python
c = Bar()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values(), gap="0%")
c.add_yaxis("商家B", Faker.values(), gap="0%")
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例", subtitle="我是副标题"))
c.render_notebook()
```

- Y 轴 formatter

```python
c = Bar()
c.add_xaxis(Faker.choose())
c.add_yaxis("商家A", Faker.values())
c.add_yaxis("商家B", Faker.values())
c.set_global_opts(title_opts=opts.TitleOpts(title="Bar-基本示例"),
                  yaxis_opts=opts.AxisOpts(
                axislabel_opts=opts.LabelOpts(formatter="{value} /月")))
c.render_notebook()
```
